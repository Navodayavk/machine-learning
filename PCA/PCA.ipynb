```python
# Importing libraries
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
np.random.seed(42)
data = pd.DataFrame({
    'feature_1': np.random.normal(size=100),
    'feature_2': np.random.normal(size=100),
    'feature_3': np.random.normal(size=100),
    'feature_4': np.random.normal(size=100),
})

scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

pca = PCA(n_components=2)
principal_components = pca.fit_transform
principal_components = pca.fit_transform(scaled_data)

pca_df = pd.DataFrame(data=principal_components,columns=['PC1','PC2'])

explained_variance = pca.explained_variance_ratio_

print("Explained Variance Ratio:",explained_variance)
print("/nPrincipal Components:")
print(pca_df.head())

plt.figure(figsize=(8, 6))
plt.scatter(pca_df['PC1'],pca_df['PC2'], c='blue' , alpha=0.7)
plt.title("PCA:Principal components")
plt.xlable("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.grid()
plt.show()
```
